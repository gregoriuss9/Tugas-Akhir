```python
# create a figure
fig = plt.figure(figsize=(14,10))
 
spec = gridspec.GridSpec(ncols=4, nrows=2,
                         width_ratios=[2,1,2,1], wspace=0.5,
                         hspace=0.3, height_ratios=[1,1,])

#################################################
ax0 = fig.add_subplot(spec[0])
#mono_tvdw 
k_m_tvdw = np.unique(data_m_tvdw[:, 0])
bands_tmvdw = np.reshape(data_m_tvdw[:, 1], (-1, len(k_m_tvdw)))

for band in range(len(bands_tmvdw)):
    ax0.plot(k_m_tvdw, bands_tmvdw[band, :]-E_m_tvdw, color = 'navy', linewidth=1.5)
    ax0.set_ylim([-5,5])
    ax0.set_xlim([min(k_m_tvdw),max(k_m_tvdw)])

k_m_vdw = np.unique(data_m_vdw[:, 0])
bands_tmvdw = np.reshape(data_m_vdw[:, 1], (-1, len(k_m_vdw)))

for band in range(len(bands_tmvdw)):
    ax0.plot(k_m_vdw, bands_tmvdw[band, :]-E_m_vdw, color = 'navy', linewidth=1.5, linestyle='dashed')
    ax0.set_ylim([-5,5])
    ax0.set_xlim([min(k_m_vdw),max(k_m_vdw)])

ax0.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax0.set_title('Monolayer')
ax0.axvline(0.6663, linewidth=0.75, color='k', alpha=0.5)
ax0.axvline(0.9982, linewidth=0.75, color='k', alpha=0.5)
ax0.set_xticks(ticks= [0, 0.6663, 0.9982, 1.57590],
                       labels=['$\Gamma$', 'K', 'M', '$\Gamma$'])
ax0.set_ylabel(r'$E-E_\mathrm{F}$ (eV)')

#################################################
ax1 = fig.add_subplot(spec[1])
ax1.plot(dos_m_tvdw, e_m_tvdw-E_m_tvdw,linewidth=1.5, color='navy')
ax1.plot(dos_m_vdw, e_m_vdw-E_m_vdw,linewidth=1.5, color='navy', linestyle='dashed')
ax1.set_ylim([-5,5])
ax1.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax1.set_xticks([])
ax1.set_xlim([0,15])
ax1.set_xlabel('DOS')

#################################################
ax2 = fig.add_subplot(spec[2])
#AA_tvdw 
k_AA_tvdw = np.unique(data_AA_tvdw[:, 0])
bands_AA_tvdw = np.reshape(data_AA_tvdw[:, 1], (-1, len(k_AA_tvdw)))

for band in range(len(bands_AA_tvdw)):
    ax2.plot(k_AA_tvdw, bands_AA_tvdw[band, :]-E_AA_tvdw, linewidth=1.5, color='darkmagenta')
    ax2.set_ylim([-5,5])

    ax2.set_xlim([min(k_AA_tvdw),max(k_AA_tvdw)])

#AA_vdw
k_AA_vdw = np.unique(data_AA_vdw[:, 0])
bands_AA_vdw = np.reshape(data_AA_vdw[:, 1], (-1, len(k_AA_vdw)))

for band in range(len(bands_AA_vdw)):
    ax2.plot(k_AA_vdw, bands_AA_vdw[band, :]-E_AA_vdw, linewidth=1.5, color='darkmagenta', linestyle='dashed')
    ax2.set_ylim([-5,5])
    ax2.set_xlim([min(k_AA_vdw),max(k_AA_vdw)])

ax2.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax2.set_title('Tumpuk AA')
ax2.axvline(0.6664, linewidth=0.75, color='k', alpha=0.5)
ax2.axvline(0.9990, linewidth=0.75, color='k', alpha=0.5)
ax2.set_xticks(ticks= [0, 0.6664, 0.9990, 1.5765], 
                     labels=['$\Gamma$', 'K', 'M', '$\Gamma$'])
ax2.set_ylabel(r'$E-E_\mathrm{F}$ (eV)')

#################################################
ax3 = fig.add_subplot(spec[3])
ax3.plot(dos_AA_tvdw, e_AA_tvdw-E_AA_tvdw,linewidth=1.5, color='darkmagenta')
ax3.plot(dos_AA_vdw, e_m_vdw-E_AA_vdw,linewidth=1.5, color='darkmagenta', linestyle='dashed')
ax3.set_ylim([-5,5])
ax3.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax3.set_xlim([0,15])
ax3.set_xticks([])
ax3.set_xlabel('DOS')

#################################################
ax4 = fig.add_subplot(spec[4])
#AB_tvdw 
k_AB_tvdw = np.unique(data_AB_tvdw[:, 0])
bands_AB_tvdw = np.reshape(data_AB_tvdw[:, 1], (-1, len(k_AB_tvdw)))

for band in range(len(bands_AB_tvdw)):
    ax4.plot(k_AB_tvdw, bands_AB_tvdw[band, :]-E_AB_tvdw, linewidth=1.5, color = 'darkgreen')
    ax4.set_ylim([-5,5])
    ax4.set_xlim([min(k_AB_tvdw),max(k_AB_tvdw)])

#AB_vdw
k_AB_vdw = np.unique(data_AB_vdw[:, 0])
bands_AB_vdw = np.reshape(data_AB_vdw[:, 1], (-1, len(k_AB_vdw)))

for band in range(len(bands_AB_vdw)):
    ax4.plot(k_AB_vdw, bands_AB_vdw[band, :]-E_AB_vdw, linewidth=1.5, color = 'darkgreen', linestyle='dashed')
    ax4.set_ylim([-5,5])
    ax4.set_xlim([min(k_AB_vdw),max(k_AB_vdw)])

ax4.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax4.set_title('Tumpuk AB')
ax4.axvline(0.6664, linewidth=0.75, color='k', alpha=0.5)
ax4.axvline(0.9990, linewidth=0.75, color='k', alpha=0.5)
ax4.set_xticks(ticks= [0, 0.6664, 0.9990, 1.5765], 
                     labels=['$\Gamma$', 'K', 'M', '$\Gamma$'])
ax4.set_ylabel(r'$E-E_\mathrm{F}$ (eV)')
#################################################
ax5 = fig.add_subplot(spec[5])
ax5.plot(dos_AB_tvdw, e_AB_tvdw-E_AB_tvdw,linewidth=1.5, color='darkgreen')
ax5.plot(dos_AB_vdw, e_m_vdw-E_AB_vdw,linewidth=1.5, color='darkgreen', linestyle='dashed')
ax5.set_ylim([-5,5])
ax5.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax5.set_xlim([0,15])
ax5.set_xticks([])
ax5.set_xlabel('DOS')

#################################################
ax6 = fig.add_subplot(spec[6])
#AC_tvdw 
k_AC_tvdw = np.unique(data_AC_tvdw[:, 0])
bands_AC_tvdw = np.reshape(data_AC_tvdw[:, 1], (-1, len(k_AC_tvdw)))

for band in range(len(bands_AC_tvdw)):
    ax6.plot(k_AC_tvdw, bands_AC_tvdw[band, :]-E_AC_tvdw, linewidth=1.5, color='darkred')
    ax6.set_ylim([-5,5])
    ax6.set_xlim([min(k_AC_tvdw),max(k_AC_tvdw)])

#AC_vdw
k_AC_vdw = np.unique(data_AC_vdw[:, 0])
bands_AC_vdw = np.reshape(data_AC_vdw[:, 1], (-1, len(k_AC_vdw)))

for band in range(len(bands_AC_vdw)):
    ax6.plot(k_AC_vdw, bands_AC_vdw[band, :]-E_AC_vdw, linewidth=1.5, color='darkred', linestyle='dashed')
    ax6.set_ylim([-5,5])
    ax6.set_xlim([min(k_AC_vdw),max(k_AC_vdw)])

ax6.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax6.set_title('Tumpuk AC')
ax6.axvline(0.6664, linewidth=0.75, color='k', alpha=0.5)
ax6.axvline(0.9990, linewidth=0.75, color='k', alpha=0.5)
ax6.set_xticks(ticks= [0, 0.6664, 0.9990, 1.5765], 
                     labels=['$\Gamma$', 'K', 'M', '$\Gamma$'])
ax6.set_ylabel(r'$E-E_\mathrm{F}$ (eV)')
#################################################
ax7 = fig.add_subplot(spec[7])
ax7.plot(dos_AC_tvdw, e_AC_tvdw-E_AC_tvdw,linewidth=1.5, color='darkred')
ax7.plot(dos_AC_vdw, e_m_vdw-E_AC_vdw,linewidth=1, color='darkred', linestyle='dashed')
ax7.set_ylim([-5,5])
ax7.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
ax7.set_xticks([])
ax7.set_xlim([0,15])
ax7.set_xlabel('DOS')

ax0.text(-0.45, 5.2, '(a)')
ax2.text(-0.45, 5.2, '(b)')
ax4.text(-0.45, 5.2, '(c)')
ax6.text(-0.45, 5.2, '(d)')

plt.savefig('bs_dos_al2s2.pdf', bbox_inches='tight')
plt.show()
```
